# Omniglot

In this notebook we analyse the Omniglot dataset

```python
import jax
import hlax
import matplotlib.pyplot as plt
```

```python
key = jax.random.PRNGKey(3141)
n_train, n_test = 100, 100
(X_train, _), (X_test, _) = hlax.datasets.load_omniglot(key, n_train, n_test, normalize=False)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(5, 5))
for i, ax in enumerate(axs.ravel()):
    ax.imshow(X_train[i].reshape(105, 105), cmap="bone")
    ax.axis("off")
plt.suptitle("Train")
plt.tight_layout()
```

```python
fig, axs = plt.subplots(3, 3, figsize=(5, 5))
for i, ax in enumerate(axs.ravel()):
    ax.imshow(X_test[i].reshape(105, 105), cmap="bone")
    ax.axis("off")
plt.suptitle("Test")
plt.tight_layout()
```
